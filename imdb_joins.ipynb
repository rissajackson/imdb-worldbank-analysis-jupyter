A query using a join to figure out what movies were made in 1965 and had over 3000 total votes to understand what was popular that year. 

```python
%%bigquery
SELECT *
FROM imdb_modeled.Movies m INNER JOIN imdb_modeled.Ratings r USING (imdb_title_id)
WHERE m.year = 1965 AND r.total_votes > 3000 LIMIT 12
```

A query using a join to figure out what movies that were considered horror movies and had a mean vote over 8. 

```python
%%bigquery
SELECT *
FROM imdb_modeled.Movies m INNER JOIN imdb_modeled.Ratings n USING (imdb_title_id)
WHERE m.genre = "Horror" AND n.mean_vote > 8 LIMIT 5
```

A query using an outer join to figure out how many recent movies from after 2015 had a director whose death year is null to test if newer movies have better data on their directors. 

```python
%%bigquery
SELECT * EXCEPT(production_company, actors, imdb_title_id)
FROM imdb_modeled.Movies m LEFT OUTER JOIN imdb_modeled.Directors n ON (m.director = n.name)
WHERE m.year > 2015 AND n.death_year IS NULL LIMIT 5
```

A query using an outer join to figure out how many writers on USA movies had a name that did not match their birth name to see how common that is. 

```python
%%bigquery
SELECT * EXCEPT(production_company, actors)
FROM imdb_modeled.Movies n RIGHT OUTER JOIN imdb_modeled.Writers m USING (imdb_title_id)
WHERE n.country = "USA" AND m.name = m.birth_name LIMIT 5
```

A query using a join to see how many writers where involved in movies that received a score greater than 5 from non US voters. This is to see how popular writers are outside of the US.

```python
%%bigquery
SELECT *
FROM imdb_modeled.Writers w JOIN imdb_modeled.Ratings d USING (imdb_title_id)
WHERE d.non_us_voters_rating > 5 LIMIT 5
```
