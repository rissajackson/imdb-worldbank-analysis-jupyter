A query using a join to figure out what movies were made in 1965 and had over 3000 total votes to understand what was popular that year. 

```python
%%bigquery
SELECT *
FROM imdb_staging.Movies m INNER JOIN imdb_staging.Ratings r USING (imdb_title_id)
WHERE m.year = 1965 AND r.total_votes > 3000 LIMIT 12
```

A query using a join to figure out what movies that were considered horror movies and had a mean vote over 8. 

```python
%%bigquery
SELECT *
FROM imdb_staging.Movies m INNER JOIN imdb_staging.Ratings n USING (imdb_title_id)
WHERE m.genre = "Horror" AND n.mean_vote > 8 LIMIT 5
```

A query using a join to figure out which people who were title principals and production designers to figure out how many people did that role. 

```python
%%bigquery
SELECT *
FROM imdb_staging.Names n INNER JOIN imdb_staging.Title_Principals t USING (imdb_name_id)
WHERE t.category = "production_designer" LIMIT 5
```

A query using an outer join to figure out how many recent movies from after 2015 had a director whose death year is null to test is newer movies have better data on their directors. 

```python
%%bigquery
SELECT * EXCEPT(production_company, actors)
FROM imdb_staging.Movies m LEFT OUTER JOIN imdb_staging.Names n ON (m.director = n.name)
WHERE m.year > 2015 AND n.death_year IS NULL LIMIT 5
```

A query using an outer join to figure out how many writers on English movies had a name that did not match their birth name to see how common that is. 

```python
%%bigquery
SELECT * EXCEPT(production_company, actors)
FROM imdb_staging.Names n RIGHT OUTER JOIN imdb_staging.Movies m ON (n.name = m.writer)
WHERE n.name != n.birth_name AND m.language = "English" LIMIT 5
```

A query using a join to see how many title principals were actors in movies that were over 180 minutes in duration to see how common longer movies are. 

```python
%%bigquery
SELECT t.imdb_title_id
FROM imdb_staging.Title_Principals t JOIN imdb_staging.Movies m ON t.imdb_title_id = m.imdb_title_id
WHERE t.category = 'actor' AND m.duration > 180 LIMIT 10
```
