Query 1: A query that joins the GDP table to the New_Movies_Beam_DF table through the Countries table using the name of each country. The join makes sure the year of a movie is the same as the year in the GDP table, and shows all the average gdp, average movie budget, and average movie worldwide income per year in Mexico.

```python
%%bigquery
select avg(gdp) as avg_gdp, avg(budget) as avg_budget, avg(worlwide_gross_income) as avg_worldwide_income, m.year
from worldbank_modeled.GDP p join worldbank_modeled.Countries c using (country_id) join imdb_modeled.New_Movies_Beam_DF m on (c.country_name = m.country)
where m.year = p.year and m.country = "Mexico" and budget is not null
group by m.year
order by m.year
```

Query 2: A query that joins the Pop_Total table to the New_Movies_Beam_DF table through the Countries table using the name of each country and then joins to the Ratings table through the imdb_title_id. The join makes sure the year of a movie is the same as the year in the Pop_Total table, and shows all the average population and the average vote for each movie produced in the United States for each year.

```python
%%bigquery
select avg(population) as avg_population, m.year, avg(mean_vote) as avg_vote 
from worldbank_modeled.Pop_Total p join worldbank_modeled.Countries c using (country_id) join imdb_modeled.New_Movies_Beam m on (c.country_name = m.country) join imdb_modeled.Ratings r using (imdb_title_id)
where m.year = p.year and m.country = "United States"
group by m.year
order by m.year
```

Query 3: A query that joins the GDP table to the New_Movies_Beam_DF table through the Countries table using the name of each country. The join makes sure the year of a movie is the same as the year in the GDP table, and shows all the average duration for each movie produced in the United States for each year and the average gdp for each year in the United States.

```python
%%bigquery
select avg(duration) as avg_duration, m.year, avg(gdp) as avg_gdp
from worldbank_modeled.GDP g join worldbank_modeled.Countries c using (country_id) join imdb_modeled.New_Movies_Beam_DF m on (c.country_name = m.country)
where m.year = g.year and m.country = "United States"
group by m.year
order by m.year
```
