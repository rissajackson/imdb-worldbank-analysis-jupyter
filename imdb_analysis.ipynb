An Aggregate query comparing the average budget for a movie per language that is also over 1000000.

```python
%%bigquery
select language, avg(budget) as avg_budget
from imdb_modeled.Movies_Beam_DF
where language != 'None'
group by language
having avg_budget > 1000000
order by language
limit 12
```

An Aggregate query for the max duration of a movie per genre that is also over 80 minutes. 

```python
%%bigquery
select genre, max(duration) as max_duration
from imdb_modeled.Movies_Beam_DF m join imdb_modeled.Actors_Beam_DF a
using (imdb_title_id)
group by genre
having max_duration > 80
order by genre
limit 12
```

An aggregate query for the average metascore for movies per year. 

```python
%%bigquery
select year, avg(metascore) as avg_metascore
from imdb_modeled.Movies_Beam_DF join imdb_modeled.Writers_Beam_DF w
using (imdb_title_id)
where metascore is not null
group by year
order by year
limit 12
```

An aggregate query for the average usa gross income compared to the average budget for movies by genre that is also over 60000 for the average USA gross income.

```python
%%bigquery
select genre, avg(usa_gross_income) as avg_usa_gross_income, avg(budget) as avg_budget
from imdb_modeled.Movies_Beam_DF  
group by genre
having avg_usa_gross_income > 60000.0
order by genre
limit 12
```

An aggregate query for the total sum of votes received per movie title where non US votes were over 1000.

```python
%%bigquery
select imdb_title_id, sum(votes_1 + votes_2 + votes_3 + votes_4 + votes_5) as total_votes
from imdb_modeled.Ratings
where non_us_voters_votes > 1000.0
group by imdb_title_id
limit 12
```

An aggregate query for the max US and non-US voter ratings for each movie writer.

```python
%%bigquery
select name, max(us_voters_rating) as max_us_rating, max(non_us_voters_rating) as max_non_us_rating
from imdb_modeled.Ratings r join imdb_modeled.Writers_Beam_DF w USING (imdb_title_id)
group by name
limit 12
```
