Movies ordered by date published that are between 70 and 120 minutes to see if there are trends in movie lengths in the mid range for duration for different decades.

```python
%%bigquery
SELECT *
FROM imdb_staging.Movies
WHERE duration > 70 AND duration < 120
ORDER BY date_published
```

Names ordered alphabetically to see how common it was for people to have between 1 and 4 divorces in the entertainment industry. 

```python
%%bigquery
SELECT *
FROM imdb_staging.Names
WHERE divorces > 0 AND divorces < 4
ORDER BY name
```

Ratings ordered by total votes with a range of mean votes between 4.0 and 6.0 to compare the total votes versus the mean votes in the mid range. 

```python
%%bigquery
SELECT *
FROM imdb_staging.Ratings
WHERE mean_vote > 4.0 AND mean_vote < 6.0
ORDER BY total_votes
```

Title principals ordered by imdb title id to see how many were dialogue writers. 

```python
%%bigquery
SELECT *
FROM imdb_staging.Title_Principals
WHERE category = "writer" AND job = "dialogue"
ORDER BY imdb_title_id
```
